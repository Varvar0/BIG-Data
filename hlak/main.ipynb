```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
engine = create_engine('postgresql+psycopg2://postgres:postgres@127.0.0.1/postgres')
```

```python
df = pd.read_sql('''SELECT Customers.id as Customers_id, Customers.full_name as customer_full_name, Customers.pasport_data,
                    contract.id as contract_id, contract.payment, contract.agrement_date, contract.expiration_date,
                    contract.PMC_id as contract_pmc_id, contract.customer_id as contract_customer_id,
                    PMC.id as PMC_id, PMC.full_name as PMC_full_name, PMC.rank, PMC.number_of_group, PMC.info_id, PMC.entry_id,
                    PMC.weapon_id, PMC.achi_id as achieviement_id, PMC.vacation_id,
                    entry_contract.id as entry_contract_id, entry_contract.date_of_entry, entry_contract.date_end_service,
                    vacations.id as vacations_id, vacations.start_date, vacations.end_date,
                    personal_info.id as personal_info_id, personal_info.medcart, personal_info.military_ticket, personal_info.bank_detail,
                    achievement.id as achievement_id, achievement.merit,
                    weapon.id as weapon_id, weapon.model,
                    Groups.id as Groups_id, Groups.title, Groups.transport_id as Groups_transport_id, 
                    Transport.id as Transport_id, Transport.model, Transport.status
                    FROM Customers
                    left JOIN contract ON Customers.id = contract.Customer_id
                    left JOIN PMC ON contract.PMC_id = PMC.id
                    left JOIN entry_contract ON entry_contract.id = PMC.entry_id
                    left JOIN vacations ON vacations.id = PMC.vacation_id
                    left JOIN personal_info ON personal_info.id = PMC.info_id
                    left JOIN achievement ON achievement.id = PMC.achi_id
                    left JOIN weapon ON weapon.id = PMC.weapon_id
                    left JOIN Groups ON Groups.id = PMC.number_of_group
                    left JOIN Transport ON Transport.id = Groups.Transport_id''', engine)
```
